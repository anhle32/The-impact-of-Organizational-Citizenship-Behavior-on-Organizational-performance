```python
Q1_H = df.CO2.quantile(0.25)
Q3_H = df.CO2.quantile(0.75)
IQR_H = Q3_H - Q1_H
H_lower_bound = Q1_H - (1.5 * IQR_H) 
H_upper_bound = Q3_H + (1.5 * IQR_H) 
df = df.drop(df[df.CO2 < H_lower_bound].index)
df = df.drop(df[df.CO2 > H_upper_bound].index)
```
